# Exercise-1

```python
import pandas as pd
import numpy as np
```

```python
#A
data=pd.read_csv("https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv",header=0)
```

```python
#B
data.columns=["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
 "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
 "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
 "peak-rpm","city-mpg","highway-mpg","price"]

#C
data.head()
```

```python
#D
data.isnull().sum()
```

```python
#E
obj_col=data.select_dtypes('object').columns
for i in obj_col:
    data[i]=data[i].replace("?",np.nan)
```

```python
data.head()
```

```python
#F
data.isnull().sum()[data.isnull().sum()>0]
```

```python
#G
data=data.dropna(subset=["price"])
```

```python
data
```

```python
data.isnull().sum()[data.isnull().sum()>0]
```

```python
#H
data[['normalized-losses','bore','stroke','horsepower','peak-rpm']]=data[['normalized-losses','bore','stroke','horsepower','peak-rpm']].astype('float64')
```

```python
data['normalized-losses'] = data['normalized-losses'].fillna((data['normalized-losses'].mean()))
data['bore'] = data['bore'].fillna((data['bore'].mean()))
data['stroke'] = data['stroke'].fillna((data['stroke'].mean()))
data['horsepower'] = data['horsepower'].fillna((data['horsepower'].mean()))
data['peak-rpm'] = data['peak-rpm'].fillna((data['peak-rpm'].mean()))
```

```python
#I
data['num-of-doors']=data['num-of-doors'].fillna(data['num-of-doors'].mode()[0],inplace=True)
```

```python
#J
data.reset_index(inplace=True)
```

```python
#K
data.dtypes
```

```python
data['price']=data['price'].astype('int64')
```

```python
data.dtypes
```

```python
#L
data['city-L/100km']=data['city-mpg']/100
data['highway-L/100km']=data['highway-mpg']/100
```

```python
#M
data['length']=data['length']/data['length'].max()
data['width']=data['width']/data['width'].max()
data['height']=data['height']/data['height'].max()
```

```python
#N
data['horsepower'].hist()
```

```python
#O and P
data['horse-binned']=pd.qcut(data['horsepower'],q=3,labels=['Low','Medium','High'])
data['horse-binned'].hist()
```

```python
#Q
col=['fuel-type','aspiration']
data=pd.concat([data.drop(col, axis=1),pd.get_dummies(data[col])],axis=1)
data.head()
```
